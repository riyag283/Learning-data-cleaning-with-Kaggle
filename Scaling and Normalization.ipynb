```python
import numpy as np
import pandas as pd

# for Box-Cox Transformation
# A Box Cox transformation is a way 
# to transform non-normal dependent variables into a normal shape.
from scipy import stats

# for min_max scaling
from mlxtend.preprocessing import minmax_scaling

# plotting modules
import seaborn as sb
import matplotlib.pyplot as plt

# set seed for reproducibility
np.random.seed(0)
```

```python
# Scaling vs. Normalization
# What is the difference?

# Both transforms the values of numeric variables 
# for the transformed data points to have specific helpful properties.

# Scaling is changing the range of data
# Normalizing is changing shape of the distribution of data.
```

```python
# Scaling fits the data in a range
# Useful when using methods that are based on measures of how far the data points are.
# E.g., SVM and KNN.
```

```python
# generate 1000 data points randomly drawn from an exponential distribution
original_data = np.random.exponential(size=1000)

# min-max scale the data between 0 and 1
scaled_data = minmax_scaling(original_data, columns=[0])

# plot both together to compare
fig, ax = plt.subplots(1,2)
sb.distplot(original_data, ax=ax[0])
ax[0].set_title("Original Data")
sb.distplot(scaled_data, ax=ax[1])
ax[1].set_title("Scaled data")
```

```python
# The point of normalization is to change your observations
# so that they can be described as a normal distribution.
```

```python
# normalize the exponential data with box-cox
normalized_data = stats.boxcox(original_data)

# plot both together to compare
fig, ax=plt.subplots(1,2)
sb.distplot(original_data, ax=ax[0])
ax[0].set_title("Original Data")
sb.distplot(normalized_data[0], ax=ax[1])
ax[1].set_title("Normalized data")
```
